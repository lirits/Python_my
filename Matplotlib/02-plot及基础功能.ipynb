```python
import matplotlib.pyplot as plt
import random
```

# 创建图像

```python
#创建画布
plt.figure()
#绘制画布
x = [1,2,3,4,5]
y = [i**2 for i in x]
plt.plot(x,y)
#显示图像
plt.show()
```

# 图像保存

```python
#创建画布
plt.figure(figsize=(20,8),dpi=100)
#绘制画布
x = [1,2,3,4,5]
y = [i**2 for i in x]
plt.plot(x,y)
# 2.1图片保存
plt.savefig('./image/02_test.png')
#显示图像
plt.show()
```

# 实例：温度图

```python
x = range(60)
y = [random.uniform(20,23) for i in x]
plt.figure(figsize=(20,8),dpi=100)
plt.plot(x,y)
plt.show()
```
